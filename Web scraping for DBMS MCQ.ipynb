# Basic Python script to collate quizes from sites using web scraping methods 

```python
#importing necessary packages 
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
```

```python
#scraping quize from javapoint 
# Making a GET request
r = requests.get('https://engineeringinterviewquestions.com/dbms-multiple-choice-questions/')
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
# parsing the html tags from various classes
ques = []
for parent in soup.find_all('div',{'class':'entry-content'}):
    for tag in parent.find_all('p'):
        for x in tag.find_all('strong'):
            for q in x:
                ques.append(q.text)

    
```
